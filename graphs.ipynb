```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('/content/drive/MyDrive/Data/Data.xlsx', 'data1')

titles = df['Title'].tolist()
counts = df['Count'].tolist()

perc = np.array(counts)
perc = np.around(perc*100/sum(perc),1)

for x in range(len(perc)):
    titles[x] += " "+str(perc[x])+"%"

plt.pie(counts, labels = titles, startangle=90)
plt.show
```
